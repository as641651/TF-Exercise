## Image processing

### Major observations:

> 1) **float** images have pixel values in the range [0,1]. So to convert datatype of images, use **tf.image.convert_image_dtype**. Using **tf.cast** will destroy the image

>2) **resizing** seems to be working only on **float images**

>3) **Color adjustment** operations (brightness, contrast etc) seems to working correctly only on **uint8** images

```python
import tensorflow as tf
import os
%matplotlib inline
import matplotlib.pyplot as plt
import glob
```

### Utilities

```python
def run_tf(x):
    
    config = tf.ConfigProto()
    config.gpu_options.allow_growth=True
    with tf.Session(config=config) as sess:    
        coord = tf.train.Coordinator()
        threads = tf.train.start_queue_runners(coord=coord)

        out = sess.run(x)

        coord.request_stop()
        coord.join(threads)
        
        return out
```

### Create data queue

```python
cat_dog_train_path = '/home/sankaran/exercise/ML/TF-Exercise/Low_Level_APIs/Cats&Dogs/train/*.jpg'
filenames = glob.glob(cat_dog_train_path)

#Queue
filename_queue = tf.train.string_input_producer(filenames,seed=0)

##Reader
reader = tf.WholeFileReader()
key,value = reader.read(filename_queue)

##Decode
image = tf.image.decode_jpeg(value)
image = tf.image.convert_image_dtype(image,dtype=tf.float32) #pixel values are in the range [0,1]
#image = tf.cast(image,tf.float32) ## This conversion breaks
#image = tf.image.resize_bicubic(tf.expand_dims(image,axis=0),(224,224)) ## works fine with tf.imgae.convert_image_dtype
image = tf.image.resize_images(image,(224,224),method=tf.image.ResizeMethod.BICUBIC) ##seems to expect float tensor
image = tf.reshape(image,[224,224,3])
```

```python
im = run_tf(image)
plt.imshow(im)
im
```

### Crop and resize

**tf.image.crop_and_resize**(

    image, 
    boxes,
    box_ind,
    crop_size,
    method='bilinear',
    extrapolation_value=0,
    name=None
)

**Args**
> **image** : 4-D Tensor [batch,h,w,c]

> **boxes** : 2-D Tensor [num boxs,4]; [**[y1,x1,y2,x2]**,...] where the co-ordinates are normalized in the range (0,1)

>**box_ind** : 1-D Tensor [num_boxes]; The image to which each box belongs; **eg :** [0,0,0,1,2,2, ...] rep 3 boxes from first image 

> **crop_size** : Size to which the return patches must be resized

```python
croped_image = tf.image.crop_and_resize(
    tf.expand_dims(image,axis=0),
    boxes=[[0,0.1,0.8,0.8],[0.5,0.5,1,1],[0.2,0.2,0.8,0.8]],
    box_ind=[0,0,0],
    crop_size=[224,224])

di = run_tf(croped_image)
for j in range(len(di)):
    plt.subplot(1, 3, j+1)
    plt.imshow(di[j, ...])
plt.show()
```

### Random processing

#### Random crop

```python
for j in range(3):
    plt.subplot(1, 3, j+1)
    distorted_image = tf.random_crop(image, [100, 100, 3])
    di = run_tf(distorted_image)
    plt.imshow(di)
plt.show()
```

#### Random flips

```python
for j in range(3):
    plt.subplot(1, 3, j+1)
    distorted_image = tf.image.random_flip_left_right(image)
    di = run_tf(distorted_image)
    plt.imshow(di)
plt.show()
```

#### Random saturation

Note: Apply on **uint8** images. 

```python
int_image = tf.image.convert_image_dtype(image,tf.uint8)
for j in range(3):
    plt.subplot(1, 3, j+1)
    distorted_image = tf.image.random_saturation(int_image,lower=0.2, upper=1.8)
    di = run_tf(distorted_image)
    plt.imshow(di)
plt.show()
```

#### Random hue

```python
int_image = tf.image.convert_image_dtype(image,tf.uint8)
for j in range(3):
    plt.subplot(1, 3, j+1)
    distorted_image = tf.image.random_hue(int_image,max_delta=0.3) # max_delta should be <= 0.5
    di = run_tf(distorted_image)
    plt.imshow(di)
plt.show()
```

### Brightness

Seems to expect **uint8** image

Delta <=0.5 seems to be the range

```python
int_image = tf.image.convert_image_dtype(image,tf.uint8)
distorted_image = tf.image.adjust_brightness(int_image,delta=0.5)
#float_image = tf.image.convert_image_dtype(distorted_image,tf.float32) ## Can be safely converted back
di = run_tf(distorted_image)
plt.imshow(di)
```

This happens with float image

```python
distorted_image = tf.image.adjust_brightness(image,delta=0.5)
di = run_tf(distorted_image)
plt.imshow(di)
```

### Contrast

Again expects **uint8**

**contrast factor** : 1 is the image itself. Slide to either side of 1 

```python
int_image = tf.image.convert_image_dtype(image,tf.uint8)
distorted_image = tf.image.adjust_contrast(int_image,contrast_factor=1.5)
float_image = tf.image.convert_image_dtype(distorted_image,tf.float32)
di = run_tf(float_image)
plt.imshow(di)
```

### Saturation

**float** image is not destroyed, but there is difference in saturation application from **uint8**

It seems safer to apply on **uint8**

**saturation factor** : 1 is the image itself. Slide to either side of 1 

```python
int_image = tf.image.convert_image_dtype(image,tf.uint8)
distorted_image = tf.image.adjust_saturation(int_image,saturation_factor=1.8)
float_image = tf.image.convert_image_dtype(distorted_image,tf.float32)
di = run_tf(float_image)
plt.imshow(di)
```

```python
distorted_image = tf.image.adjust_saturation(image,saturation_factor=1.8)
di = run_tf(image)
plt.imshow(di)
```
